```python
###import libraries

import gym
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
###prepare environment

env = gym.make('MountainCar-v0')
env.seed(1)
env.reset()
```

```python
###understand environment

"""
Description: 
    - The goal is to drive up the mountain on the right (position of 0.5)
    - The car's engine is not strong enough to scale the mountain in a single pass
    - The only way to succeed is to drive back and forth to build up momentum

Observation: 
    Type: Box(2)
    Num	Observation      Min         Max
    0	Position        -1.2         0.6
    1	Velocity        -0.07        0.07

Actions:
    Type: Discrete(3)
    Num	Action
    0	Push left
    1	No push
    2	Push right

Reward:
    - Reward is -1 for every step taken, including the termination step (whether the goal is reached or not)

Starting State:
    - Random position from -0.6 to -0.4 with no velocity

Episode Termination:
    - Position reaches 0.5
    - Episode length is greater than 200
"""
```

```python
###visualize environment

for episode in range(3):
    print('Episode', episode+1)
    env.reset()
    done = False
    while not done:
        env.render()
        action = env.action_space.sample()
        observation, reward, done, info = env.step(action)
        print('\tAction\t\t:', action)
        print('\tObservation\t:', observation)
        print('\tReward\t\t:', reward)
        print('\tDone\t\t:', done)
        print('\tInfo\t\t:', info, '\n')

env.close()
```

```python
###prepare model for training

class DPGAgent:
    def __init__(self, observation_size, action_size):
        #get observation and action sizes
        self.observation_size = observation_size
        self.action_size = action_size
        #set hyperparameters
        self.alpha = 0.01
        self.gamma = 0.95
        #initialize variable arrays
        self.observation_array, self.action_array, self.reward_array = [], [], []
        #build model
        self.build_model()
      
    def build_model(self):
        with tf.name_scope('inputs'):
            self.tf_observations = tf.placeholder(tf.float32, [None, self.observation_size], name='observations') #tensor for observations
            self.tf_actions = tf.placeholder(tf.int32, [None, self.action_size], name='num_actions') #tensor for actions
            self.tf_values = tf.placeholder(tf.float32, [None, ], name='observation_values') #tensor for observation values (discounted and normalized rewards)
        
        with tf.name_scope('fc1'):
            fc1 = tf.contrib.layers.fully_connected(self.tf_observations, #observation tensor is fed into input layer
                                                    num_outputs=16, #first hidden layer has 16 nodes
                                                    activation_fn=tf.nn.tanh)
        
        with tf.name_scope('fc2'):
            fc2 = tf.contrib.layers.fully_connected(fc1, 
                                                    num_outputs=32, #second hidden layer has 32 nodes
                                                    activation_fn=tf.nn.tanh)
        
        with tf.name_scope('fc3'):
            fc3 = tf.contrib.layers.fully_connected(fc2,
                                                    num_outputs=self.action_size, #output layer has 2 nodes
                                                    activation_fn=None) #no special activation function, just a linear activation function that decides between 0, 1, or 2
        
        self.action_distribution = tf.nn.softmax(fc3)
        with tf.name_scope('loss'):
            neg_log_prob = tf.nn.softmax_cross_entropy_with_logits_v2(logits=fc3, labels=self.tf_actions)
            self.loss = tf.reduce_mean(self.tf_values*neg_log_prob)
        
        with tf.name_scope('train'):
            self.optimizer = tf.train.AdamOptimizer(self.alpha).minimize(self.loss)
    
    def act(self, observation):
        #action probabilities given the observation
        action_probability_distribution = sess.run(self.action_distribution, feed_dict={self.tf_observations: observation})
        #randomly choose action according to probabilities
        return np.random.choice(range(action_probability_distribution.shape[1]), p=action_probability_distribution.ravel())

    def downhill(self, new_position, velocity, new_velocity, downhill_from_left, downhill_from_right):
        if new_position < -0.5 and new_velocity > 0: downhill_from_left = True
        if new_position > -0.5 and new_velocity < 0: downhill_from_right = True
        if ((velocity < 0 and new_velocity > 0) or 
            (velocity > 0 and new_velocity < 0)):
            downhill_from_left = False
            downhill_from_right = False
        return downhill_from_left, downhill_from_right

    def modify_reward(self, action, new_position, new_velocity, downhill_from_left, downhill_from_right):
        push_left = action == 0
        push_right = action == 2
        speed = np.abs(new_velocity)
        modified_reward = 0
        position_reward = np.abs(new_position - (-0.5))
        if new_position > 0.4: position_reward += 250
        speeding_reward = 0
        if ((downhill_from_left and push_right) or 
            (downhill_from_right and push_left)):
            speeding_reward = 250 * speed
        modified_reward = position_reward + speeding_reward
        return modified_reward, position_reward, speeding_reward

    def remember(self, observation, action, reward): #the 21st night of September
        self.observation_array.append(observation)
        action_hot_encoded = np.zeros(self.action_size)
        action_hot_encoded[action] = 1
        self.action_array.append(action_hot_encoded)
        self.reward_array.append(reward)
    
    def some_stats(self, new_position, furthest_position, new_velocity, highest_speed):
        if new_position > furthest_position: furthest_position = new_position
        if np.abs(new_velocity) > highest_speed: highest_speed = np.abs(new_velocity)
        return furthest_position, highest_speed

    def discount_and_normalize(self):
        observation_values = np.zeros_like(self.reward_array)
        cumulative = 0.0
        for i in reversed(range(len(self.reward_array))):
            cumulative = cumulative*self.gamma + self.reward_array[i]
            observation_values[i] = cumulative
        
        mean = np.mean(observation_values)
        std = np.std(observation_values)
        observation_values = (observation_values - mean) / std
        return observation_values

    def print_text(self, episode, G, modified_G, position_G, speeding_G, furthest_position, highest_speed):
        print('Episode', episode+1)
        print('\tScore\t\t\t:', G)
        print('\tModified score\t\t:', modified_G)
#        print('\tPosition score\t\t:', position_G)
#        print('\tSpeeding score\t\t:', speeding_G)
        print('\tFurthest position\t:', furthest_position)
        print('\tHighest speed\t\t:', highest_speed, '\n')

    def clear(self): #clear memory after the end of each episode
        self.observation_array, self.action_array, self.reward_array = [], [], []
```

```python
## train model

observation_size = env.observation_space.shape[0]
action_size = env.action_space.n
agent = DPGAgent(observation_size, action_size)
G_array = []
modified_G_array = []
furthest_position_array = []
highest_speed_array = []
loss_array = []
highest_G = -200.0
highest_modified_G = 0
n_episodes = 1000
saver = tf.train.Saver()
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for episode in range(n_episodes):
        observation = env.reset()
        observation = np.reshape(observation, [1, observation_size])
        position = observation[0][0]
        velocity = observation[0][1]
        G = 0
        modified_G = 0
        position_G = 0
        speeding_G = 0
        furthest_position = -1.2
        highest_speed = 0
        downhill_from_left = False
        downhill_from_right = False
        while True:
            #env.render()
            action = agent.act(observation)
            new_observation, reward, done, info = env.step(action)
            new_observation = np.reshape(new_observation, [1, observation_size])
            new_position = new_observation[0][0]
            new_velocity = new_observation[0][1]
            downhill_from_left, downhill_from_right = agent.downhill(new_position, velocity, new_velocity, downhill_from_left, downhill_from_right)
            modified_reward, position_reward, speeding_reward = agent.modify_reward(action, new_position, new_velocity, downhill_from_left, downhill_from_right)
            agent.remember(new_observation, action, modified_reward)
            furthest_position, highest_speed = agent.some_stats(new_position, furthest_position, new_velocity, highest_speed)
            G += reward
            modified_G += modified_reward
            position_G += position_reward
            speeding_G += speeding_reward
            observation, position, velocity = new_observation, new_position, new_velocity
            if done:
                observation_values = agent.discount_and_normalize()
                loss, _ = sess.run([agent.loss, agent.optimizer], feed_dict={agent.tf_observations: np.vstack(agent.observation_array),
                                                     agent.tf_actions: np.array(agent.action_array),
                                                     agent.tf_values: observation_values})
                agent.print_text(episode, G, modified_G, position_G, speeding_G, furthest_position, highest_speed)
                if G > highest_G or G > (np.mean(G_array) + 2*np.std(G_array)):
                    saver.save(sess, './models/model.ckpt')
                    print('Model saved')
                if G > highest_G: highest_G = G
                if modified_G > highest_modified_G: highest_modified_G = modified_G
                G_array.append(G)
                modified_G_array.append(modified_G)
                furthest_position_array.append(furthest_position)
                highest_speed_array.append(highest_speed)
                loss_array.append(loss)
                agent.clear()
                break

print('Highest score\t\t:', highest_G)
print('Highest modified score\t:', highest_modified_G)
env.close()
```
